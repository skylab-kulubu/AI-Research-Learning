Supervised Learning 
x(input) -> y(output)
In supervised learning,we provide inputs and their corresponding correct answers(outputs) to the algorithm for laerning.Subsequently, we can present the algorithm with new input values (x) to predict their corresponding output values(y).

Summary :
Supervised Learning'de bazı x değerleri için y değerlerini biliyoruz. Bu x değerleri ve y değerleri ile algoritmayı eğitiyoruz. Böylece algoritma x ile y değerleri arasında bağlantı kuran bir fonksiyon buluyor. Daha sonrasında y değerlerini bilmediğimiz x değerlerini bu fonksiyona yazarak belirli bir hata oranıyla x için y değerini tahmin etmiş oluyoruz.

Unsupervised Learining
In unsupervised algorithms, the algorithm is given a dataset without explicit instructions on what to do.
The algorithm autonomously identifies patterns, relationships, or structures in the data.

Example :
Given a set of news articles found on the web,group them into sets of articles about the same story

Gradient Descent
Gradient Descent verilen bir başlangıç noktası için fonksiyonun türevinin (gradyanı) tersi doğrultusunda azalarak minimum değere iniş yapmamız demektir.
Denklemi :
x_new = x_old -eps*df
Başlangıçtaki denklemde yani x_new=x_old-eps*sign(df) dediğimiz zaman türevi sadece x_old değerinden doğru yönde ilerlemek için kullanmış oluyoruz. ama denklemi şu şekilde revize edersek x_new = x_old -eps*df olarak yenilersek  x_old değerinden her iterasyonda df*eps ifadesini çıkarmış oluruz . fonksiyon üzerinde minimuma türevin 0 olduğu yere ilerlediğimiz için türev değeri de azalmakta böylelikle türevin büyük olduğu yerlerde daha büyük adımlar atarak türevin küçük olduğu yerlerde daha küçük adımlar atarak minimum değere yaklaşmış oluyoruz . bu da iterasyon sayımızı azaltmış oluyor

Overfitting
Model x^2 iken x^4 kullanmamız.
Modelimiz overfit olması modeli fazla eğitmemiz ve neredeyse her veri noktasından geçecek hale getirmemiz gibi düşünülebilir. Model elimizdeki verilere o kadar adapte olmuştur ki yeni bir veri eklediğimizde model onu da kapsayacak şekilde değişir.
Model veri kümesindeki her y değerinden geçmeye çalışır. Bu da genel bir path yakalamak yerine verileri ezberlemesine sebep oluyor. Overfitting'te modelimiz fazla esnek.

Underfitting
Model x^4 iken x^2 kullanmamız.
Modelimizin underfitting olması modelin az eğitilmesi olarak düşünülebilir .Model yeterince eğitilmediği için tahminlerimiz yanlış gelir.Modelimiz değişikliklere iyi adapte olamamıştır .Underfitting'te modelimiz fazla katı.
