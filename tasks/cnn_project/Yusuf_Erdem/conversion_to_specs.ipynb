```python
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
import librosa
import librosa.display
import numpy as np
import os
```

```python
img_shape = (100, 100)
def create_spectogram(y):
    spec = librosa.feature.melspectrogram(y=y)
    spec = librosa.amplitude_to_db(spec, ref=np.max)
    spec_normalized = (spec - spec.min()) / (spec.max() - spec.min()) * 255
    fin = spec_normalized.astype(np.uint8)
    fin = cv2.resize(fin, img_shape )
    return fin

```

```python
with open(r"metadata/UrbanSound8K.csv", "r", encoding="utf-8") as f:
    data = f.readlines()


#Aşağıda yapacağımız şey:
#   metadata'daki csv'yi aç, her satırdaki resmi bul, spektogramını çıkar, uygun klasöre gidip kaydet.
```

```python
hist = np.zeros(10, dtype=np.int32)
for i, line in enumerate(data[1:]):             #remember i starts from 0
    elements = line.rsplit()[0].split(",")
    fold = elements[5]
    classID = int(elements[6])
    path = "audio/fold" + fold + "/" + elements[0]
    y, sr = librosa.load(path)
    spec = create_spectogram(y)
    
    directory = "specs/class" + str(classID)
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    path = directory + "/" + str(hist[classID]) + ".png"
    cv2.imwrite(path, spec)
    hist[classID] += 1
    
    print(f"line {i+1} is done. hist[classID] = {hist[classID]}, classID = {classID}")
    
    
```
